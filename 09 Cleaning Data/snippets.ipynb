```python
import pandas as pd
import numpy as np
```

```python
people = {
    'first': ['Karthik', 'Jane', 'John', 'Chris', np.nan, None, 'NA'],
    'last': ['Raja', 'Doe', 'Doe', 'Raja', np.nan, np.nan, 'Missing'],
    'email': ['KarthikRaja@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
df = pd.DataFrame(people)
df
```

```python
df.dropna()  # df.dropna(axis = 0, how = 'any', subset=None)
```

```python
df.dropna(axis=0 , how='all')   # drops only if all the values are missing
```

```python
df.dropna(axis='columns' , how='all')
```

```python
df.dropna(axis='columns' , how='any')
```

```python
df.dropna(axis=0 , how='all', subset=['email'])
```

```python
df.dropna(axis=0 , how='any', subset=['email'])  # how has no work here
```

```python
df.dropna(axis='index', how='all', subset=['last','email'])
```

```python
df = pd.DataFrame(people)

df.replace('Na', np.nan, inplace=True)
df.replace('Missing', np.nan, inplace=True)
```
